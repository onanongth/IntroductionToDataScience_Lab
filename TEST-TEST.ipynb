```python
A = np.matrix(np.linspace(0,100))
print(A)
Y = np.matrix(142.323+40.029*A+np.exp(-0.00684*A))
print(Y)
CdPinv = np.linalg.pinv(A)
beta_bdPinv = np.matmul(CdPinv,Y)
print(beta_bdPinv)
print(A*beta_bdPinv)
```

```python
u = np.array([[1],[2],[3],[4]])
v = np.array([[-2],[1],[1]])
vt = np.transpose(v)
print(u.shape)
print(v.shape)
print('an output matrix of the outer production = ',u.shape[0],'*',vt.shape[1])
```

```python
outUV = np.zeros((u.shape[0],vt.shape[1]))
rowU = u.shape[0]
colU = u.shape[1]
colVT = vt.shape[1]
for i in range(rowU):
    for j in range(colVT):
        for k in range(colU):
            outUV[i][j] += u[i][k] *vt[k][j]
            
print(outUV)
```
